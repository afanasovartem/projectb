###Импорт необходимых библиотек

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

###Загрузка и первичный обзор

```python
df = pd.read_csv('google_books_dataset.csv', encoding = 'latin1')
df.head() #первые 5 строк
```

```python
df.tail()#последние 5 строк
```

```python
#структура данных
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns #столбцы
```

```python
df['published_date'] = pd.to_datetime(df['published_date'], errors='coerce', utc=True)
df
```

```python
df['isbn_13'] = pd.to_numeric(df['isbn_13'], errors='coerce')
df
```

```python
df['isbn_10'] = pd.to_numeric(df['isbn_10'], errors='coerce')
df
```

###Обработка данных

```python
df.isnull().sum() #количесвто пропусков
```

```python
for name in df.select_dtypes(include=['number']).columns.tolist():
  df[name].fillna(df[name].median(), inplace = True)
df
```

```python
for name in df.select_dtypes(include=['object']).columns.to_list():
  df[name].fillna('Unknown', inplace=True)
df
```

```python
df['published_date'] = df['published_date'].fillna(df['published_date'].median())
df
```

###Дубликаты

```python
duplicates = df.duplicated()
print("Найдено дубликатов:", duplicates.sum())
```

###Выбросы

```python
sns.boxplot(data=df, y='page_count')
plt.title('Распределение количества страниц', fontsize=15)
plt.ylabel('Страницы')
plt.show()
```

###Визуализация

```python
df['page_count'].hist(bins=50, color='blue')#столбчатая диограмма
plt.title('число страниц по книгам')
plt.xlabel('колличестов страниц')
plt.ylabel('Количество книг')
```

```python
plt.figure(figsize=(8,4)) #гистограмма распределения языков - почти все книги на английском
df['language'].value_counts().head(10).plot(kind='bar')
plt.title('Топ-10 языков книг')
plt.xlabel('Язык')
plt.ylabel('Количество')
plt.show()
```

```python
#Линейный график
plt.figure(figsize=(10,4))
df.groupby('published_date').size().plot()
plt.title('Количество книг по годам')
plt.xlabel('Год')
plt.ylabel('Количество книг')

```

```python
df['authors'].value_counts().head(10).plot.pie()
```

```python
popular_autors = df['authors'].value_counts().head(10).index.tolist()
new_df = df[df['authors'].isin(popular_autors)]
new_df
```

```python
sns.boxplot(data=new_df, x='authors', y='list_price')
plt.title('Распледеление цен книг по авторам', fontsize = 15)
plt.xlabel('автор')
plt.ylabel('цена')
plt.xticks(rotation = 67)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

###Группировки и агрегации

```python
group1 = df.groupby(['language'])['page_count'].mean().sort_values()
print("Средние страницы по языкам:")
group1
```

```python
group2 = df.groupby(['authors'])['average_rating'].mean()
print("Средний рейтинг по авторам:")
group2
```

```python
group3 = df.groupby('authors')['list_price'].sum()
print("Сумма цен всех книг по авторам:")
group3
```
